```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data=pd.read_csv("churn_Modelling.csv")
data
```

```python
#preprocess the data
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
#encode categorical variable
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
```

```python
data
```

```python
#OHE
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_geo=OneHotEncoder(sparse_output=False)
geo_encoder=one_hot_geo.fit_transform(data[['Geography']])
```

```python
geo_encoder
```

```python
one_hot_geo.get_feature_names_out(['Geography'])
```

```python
geo_df=pd.DataFrame(geo_encoder,columns=one_hot_geo.get_feature_names_out(['Geography']))
```

```python
# Concatenate with original data
data = pd.concat([data.drop('Geography',axis=1), geo_df], axis=1)
```

```python
data.head()
```

```python
#save the encoders and scalers
with open('label_encoder_gender.pkl','wb') as file :
    pickle.dump(label_encoder_gender,file)
with open('onehot_encoder.pkl','wb') as file :
    pickle.dump(one_hot_geo,file)
     
```

```python
#diviide the data set into dependent and independent feature
X=data.drop('Exited',axis=1)
y=data['Exited']

#split the data into train and test set
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
# Convert all inputs to float32
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
y_train = y_train.astype('float32')
y_test = y_test.astype('float32')

#scale these features
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
##ANN Implementation
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

import datetime
```

```python
# Build ANN model
model = Sequential([
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),  # HL1 connected to input layer
    Dense(32, activation='relu'),                                     # HL2
    Dense(1, activation='sigmoid')                                     # Output layer
])
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
#compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
##set up the tensorboard
log_dir = "logs/fit_" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
##setup early stopping
early_stopping_callback=EarlyStopping(monitor='val_loss',
                                      patience=5,
                                      restore_best_weights=True)
```

```python
history = model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=100,
    callbacks=[tensorboard_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```
